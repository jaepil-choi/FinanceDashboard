```python
from pathlib import Path

import pandas as pd
```

```python
cwd = Path('.').resolve()
data_path = cwd / 'data'
```

```python
df = pd.read_pickle(data_path / 'allAssets_df.pkl')
df
```

```python
df.columns
```

```python
assets = list(df.columns)
assetclasses = ['E', 'F', 'C', 'B'] # E: 'equity', F: 'fx', C: 'commodity', B: 'bond'

assetclass2asset = {
    'E': assets[0:9],
    'F': assets[9:17],
    'C': assets[17:26],
    'B': assets[26:]
}
```

```python
asset2assetclass = {}

for asstcls, assts in assetclass2asset.items():
    for asst in assts:
        asset2assetclass[asst] = asstcls
```

```python
def asset2code(asset):
    assetclass = asset2assetclass[asset]
    assetindex = assets.index(asset)

    return assetclass + str(assetindex)
    
```

```python
df.columns = [asset2code(col) for col in df.columns]
```

```python
df.to_pickle(data_path / 'crossasset.pkl')
```
